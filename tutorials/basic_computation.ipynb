We import the contents of the package

```julia
using FractionalKrylovMC, LinearAlgebra
```

We define a problem

```julia
k = 20; # size of matrix
A = rand(k,k); Q = Matrix(qr(A).Q)
A = Q*diagm(rand(k))*Q'
t = rand()*2
α = 0.1 + rand()*0.9; # time fractional coefficient
γ = 0.1 + rand()*0.9; # space fractional coefficient
α = 0.9; γ = 1. # I need to test integral formula with γ bigger than α
u0 = rand(k)*10 # initial condition
problem = MittagLefflerProblem(A,u0,t,α,γ)
```

And we call the solution with two different solvers

```julia
cutoff = 1e7 # ignoring all times above the cutoff
nsims = Int(1e5)
MCSol = solve(problem,nsims,cutoff,MCSolver())
```

```julia
MCSol.uT
```

```julia
EigenSol = solve(problem,EigenSolver())
@show EigenSol.uT
@show MCSol.uT
```
